```python
#Libreriass

import pandas as pd
import numpy as np
import seaborn as sns
import math
```

```python
#Base de datos

url= 'https://raw.githubusercontent.com/ousuga/Proyecto1_IAN/main/precios_pdtos_agricolas.cvs'
df = pd.read_csv(url) #creacin del data frame
df
```

```python
'''
COMO EL DATA FRAME PRESENTA DATOS NULOS COMO n.d. y Nos interesa tenerlo
como NaN (Nulos) debo realizar ese cambio en de Data frame
'''
#REEMPLAZAR N.D. A NULOS

df['precio'] = df['precio'].replace('n.d.', np.nan)
df

df['variabilidad'] = df['variabilidad'].replace('n.d.', np.nan)
df
```

```python
'''
1RO: Procedo a indentificar los datos duplicados de diferentes formas para
analizar que puedo hacer con ellos
'''

#DATOS DUPLICADOS

# El uso del keep='first' considera el primer valor como nico y los dems como duplicados
duplicados = df.duplicated(keep='first')

# Imprimir las filas duplicadas
print(duplicados)

```

```python
# Identificar filas duplicadas
duplicados = df.duplicated()

# Imprimir las filas duplicadas, mostrando cada columna para analizar detalladamente
print(df[duplicados])
```

```python
#Eliminar datos duplicados
'''
Eliminar los datos duplicados para mantener la integridad del analsis y
evitar la distorcin de la base de datos, obteniendo com resultado un
modelo ms preciso
'''

df=df.drop_duplicates()

df
```

3ERO: Luego de ver y analizar que los datos faltantes son una cantidad
considerable del data frame por lo que no es considerable eliminarlos
ya que me puede alterar mi analsis, por lo tanto se procede a imputarlos
de una manera adecuada

- Puede ser con alguna medida descriptiva (media, moda o mediana)
- llenar el dato faltante con el anterior (esta no es la medida adecuada
porque los producto son muy diferentes entre si por lo que no van en un orden secuencial de
menor a mayor o se podra organizar el data frame segn la columna de precio de mayor a menor
y llenarlo con el dato anterior)
- Interpolar no es el mejor metodo para realizar esto ya que puede que tenga un valor
diferente de un producto en otra fecha y no sera muy congruente

- Procedo a llenar los valores faltantes con la mediana del precio



```python
#Sacamos la mediana (mediana porque no es tan afectada por los valores atipicos)

mediana = df['precio'].median()
mediana
```

```python
#Reemplazar la mediana obtenida (2.775) en los datos de precio nulos
df['precio'] =df['precio'].fillna(mediana)
df
```

```python
# Verificar que columnas tienen nulos

df['precio'].isnull().sum() #est lista la columna de precios
```

```python
#convertir la variabilidad que esta como object a float

df['variabilidad'] = pd.to_numeric(df['variabilidad'], errors='coerce', downcast='float')


df.info()
```

```python
#Ahora a seguir llenando las columnas que tengan datos nulos -> Variabilidad

'''
Decido reemplazar la variabilidad utilizando una interpolacin lineal
esto me ayudar a tener en cuenta el dato el dato anterior y el siguiente.
Esto me permitir "suavizar" los datos faltantes
'''

# Interpola un valor numerico con el mtodo lineal
#df.interpolate(method='linear')

df['variabilidad'].interpolate(method='linear', inplace=True)

df
```

```python
#4 Cual es el producto ms caro en promedio en todas las ciudades?

# Agrupar los datos por producto y calcula el precio promedio para cada uno
precio_promedio_por_producto = df.groupby('producto')['precio'].mean() #groupby se utiliza para agrupar los datos por colimna


# Encontrar el producto con el precio promedio ms alto
producto_mas_caro = precio_promedio_por_producto.idxmax() #idmax producto con la media de precio ms alta
precio_mas_caro = precio_promedio_por_producto.max()

producto_mas_caro
```

```python
#5 Que fecha se registr el precio mas alto para cada producto en cada ciudad?

# Agrupar los datos por ciudad y producto
grupos = df.groupby(['ciudad', 'producto'])

# Crea una lista para almacenar los resultados
resultados = []

# ciclo para cada grupo
for nombre, grupo in grupos:
    fila_max_precio = grupo.loc[grupo['precio'].idxmax()] #fila con el precio ms alto en cada grupo
    resultados.append(fila_max_precio)

# Data frame nuevo
resultados = pd.DataFrame(resultados)

# Precio ms alto por cada ciudad y producto
resultados[['ciudad', 'producto', 'fecha', 'precio']]


#Nota: Hay das diferentes en los que un producto cost lo mismo y quise mantener
#solo el primer dato con un keep = first pero no fu capaz :(
#De todas formas el data frame funcion

```
